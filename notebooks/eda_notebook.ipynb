```python
import pandas as pd
```

```python
df = pd.read_csv("../data/GDPa1_246 IgGs_cleaned.csv")
```

```python
df
```

```python
# Human-readable identifiers for each antibody candidate.
# These are kept for clarity during analysis and debugging but will NOT be used as model features.
identifier_cols = [
    "antibody_id",      # Unique internal identifier for the antibody candidate
    "antibody_name"     # Human-readable name or label for the antibody
]

# Biophysical and developability assays used as predictive features.
# These measurements describe stability, aggregation propensity, hydrophobicity,
# and polyspecificity of the antibody, which can influence expression yield (titer).
assay_cols = [
    "SMAC",             # Self-interaction measurement indicating aggregation tendency
    "HIC",              # Hydrophobic Interaction Chromatography score (hydrophobicity)
    "HAC",              # Hydrophobic interaction metric related to surface hydrophobic patches
    "PR_CHO",           # Polyspecificity reagent binding (CHO cell proteins)
    "PR_Ova",           # Polyspecificity reagent binding (ovalbumin control)
    "AC-SINS_pH6.0",    # Self-interaction assay measuring aggregation propensity at pH 6.0
    "AC-SINS_pH7.4",    # Self-interaction assay measuring aggregation propensity at physiological pH
    "Tonset",           # Temperature at which unfolding begins (thermal stability)
    "Tm1",              # First melting temperature transition (protein stability)
    "Tm2",              # Second melting temperature transition (additional stability transition)
]

# Protein sequence inputs used to generate embeddings with protein language models (PLMs).
# These represent the antibody variable regions responsible for antigen binding.
# The sequences themselves are not used directly as features but are converted into
# numerical embeddings using models such as ESM-C, ProtT5, ANKH, etc.
sequence_cols = [
    "vh_protein_sequence",  # Variable region amino acid sequence of the heavy chain
    "vl_protein_sequence",  # Variable region amino acid sequence of the light chain
]

# Categorical descriptors of the antibody chains.
# These represent antibody structural subclasses and will be encoded during modeling.
categorical_cols = [
    "hc_subtype",       # Heavy chain subtype (e.g., IgG1, IgG2)
    "lc_subtype",       # Light chain type (e.g., kappa or lambda)
]

# Target variable for prediction.
# Titer represents the expression yield of the antibody in the production system
# and is the primary developability metric we aim to predict.
target = "Titer"
```

```python
filtered_cols = (identifier_cols + assay_cols + sequence_cols + categorical_cols + [target])

df_filtered = df[filtered_cols].copy()
```
